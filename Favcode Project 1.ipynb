```python
#importing the necessary libraries for fecthing the data from the site  
import pandas as pd
import requests 
from bs4 import BeautifulSoup as bs
from datetime import datetime
import re
raw_html = requests.get(url="http://www.nairaland.com/home")
raw_data = raw_html.text
soup_data = bs(raw_data, "lxml")
soup_data("td")
```

```python
#using the "for in" method to extract the text from the html
for data in soup_data("td"):
    print(data.text)
```

```python
#extracting the birthday of the users for today
member_found = None
re_match = "[\w]+\([\d]+\)"
for data in soup_data("td"):
    data_found = re.findall(re_match, data.text)
    
    if data_found:
        member_found = data_found

print (member_found)
```

```python
member_found_replaced = [x.replace(")", "") for x in member_found]            

print (member_found_replaced)
```

```python
for y in member_found_replaced:
    member_cleaned = y.split("(")
    print (member_cleaned)
```

```python
member_cleaned = {}

for y in member_found_replaced:
    temp_data = y.split("(")
    
    member_cleaned[temp_data[0]] = int(temp_data[1])
    
print (member_cleaned)
```

```python
#converting the scrapped data into a data frame
columns_name = ["Username", "Age"]
df = pd.DataFrame(list(member_cleaned.items()), columns = columns_name )

df
```
